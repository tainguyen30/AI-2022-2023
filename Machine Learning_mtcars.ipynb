```python
import pandas as pd
from matplotlib import pyplot as plt

# mpg = mies per gallon, disp = displacement: dung tích buồng đốt
# hp = gross horsepower: số ngựa, wt = weight (lb/1000): khối lượng xe
url = "mtcars.csv"
data = pd.read_csv(url)
data.hist()   # Sự lặp lại của dữ liệu
plt.show

correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind="scatter", x="wt", y="mpg", figsize=(9,9), color="black")
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data["wt"]), y=data["mpg"])
# y = ax + b (a = coefficient: hệ số góc, b = intercept: giao tuyến)
print(model.intercept_)
print(model.coef_)
model.score(X=pd.DataFrame(data["wt"]), y=data["mpg"])    # precision: độ chính xác
```

```python
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black', xlim = (0,7))
y_pred = model.predict(X = pd.DataFrame(data['wt']))
plt.plot(data['wt'],y_pred,color='blue')
```

```python
# Hàm bậc 2
import numpy as np
model = linear_model.LinearRegression()
X = pd.DataFrame([data["wt"], data["wt"]**2]).T
model.fit(X, y=data["mpg"])
print("Intercept of the model: ", model.intercept_)
print("Cofficient of the model: ", model.coef_)
print("The precision of the model: ", model.score(X, y=data["mpg"]))

range = np.arange(1.5, 5.5, 0.1)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black', xlim = (0.7))
prediction = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X=prediction)
plt.plot(range, y_pred, color="blue")
```

```python
# Hàm bậc 3
model = linear_model.LinearRegression()
X = pd.DataFrame([data["wt"], data["wt"]**2, data["wt"]**3]).T
model.fit(X, y=data["mpg"])
print("Intercept of the model: ", model.intercept_)
print("Cofficient of the model: ", model.coef_)
print("The precision of the model: ", model.score(X, y=data["mpg"]))

range = np.arange(1.5, 5.5, 0.1)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black', xlim = (0.7))
prediction = pd.DataFrame([range, range**2, range**3]).T
y_pred = model.predict(X=prediction)
plt.plot(range, y_pred, color="blue")
```

```python
model = linear_model.LinearRegression()
model.fit(X=data.loc[:, {"wt", "hp"}], y=data["mpg"])
print(model.intercept_)
print(model.coef_)
print(model.score(X=data.loc[:, {"wt", "hp"}], y=data["mpg"]))
```
